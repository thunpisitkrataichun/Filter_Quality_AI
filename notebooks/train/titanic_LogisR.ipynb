LR หรือ Logistic Regression เป็นโมเดลสำหรับทำนายผลออกมาเป็นในเชิง True false เช่น คุณเป็นมะเร็งหรือไม่ คุณมีเเนวโน้มที่จะเป็นโควิดหรือไม่

```python
import pandas as pd
import numpy as np
import joblib 
import torch
import torch.nn as nn

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_absolute_error , r2_score , accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score , f1_score , recall_score , precision_score
from sklearn.discriminant_analysis import StandardScaler
```

```python
#Load Dataset

dataset = pd.read_csv("../../data/processed/Titanic-Dataset.csv")
```

```python
target = "Survived"

x = dataset.drop(columns=[target, "PassengerId", "Name", "Ticket"])

# Cabin
x["Cabin"] = x["Cabin"].str[0]
x["Cabin"] = x["Cabin"].fillna("Unknown")

# Sex
x["Sex"] = x["Sex"].map({"male": 1, "female": 0})

# Embarked
x["Embarked"] = x["Embarked"].fillna("Unknown")
x = pd.get_dummies(x, columns=["Embarked", "Cabin"])

# Age
x["Age"] = x["Age"].fillna(x["Age"].median())

# fill numeric
x = x.fillna(x.mean(numeric_only=True))

# convert type
x = x.astype(float)

y = dataset[target]

print(x.head())
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)

x_train , x_test , y_train , y_test = train_test_split( x , y, test_size=0.2)

model = LogisticRegression(max_iter=1000    )

model.fit(x_train , y_train)

y_pred = model.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred)) # ทายถูกกี่ %
print("Precision:", precision_score(y_test , y_pred)) # ทายว่า “รอด” แล้วถูกจริงกี่ %
print("Recall:", recall_score(y_test, y_pred))  # คนที่ “รอดจริง” เราหาเจอกี่ %
print("F1:", f1_score(y_test, y_pred)) # balance ระหว่าง precision กับ recall
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, x, y, cv=500)
print(scores)
print(scores.mean())
```

```python
joblib.dump(model , "titanic_LR.pkl")
```
